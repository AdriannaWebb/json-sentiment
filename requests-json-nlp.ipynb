### 2. Read in the contents of your file.  Print the lyrics of the song (not the entire dictionary!) and use spaCyTextBlob to perform sentiment analysis on the lyrics.  Print the polarity score of the sentiment analysis.  Given that the range of the polarity score is `[-1.0,1.0]` which corresponds to how positive or negative the text in question is, do you think the lyrics have a more positive or negative connotaion?  Answer this question in a comment in your code cell.

```python
import json
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

# Read the JSON file from Question 1
with open("poem_data.json", "r") as f:
    poem_data = json.load(f)

# Get just the poem text (not the entire dictionary)
poem_text = '\n'.join(poem_data[0]['lines'])
print("Poem:")
print(poem_text)

# Set up spaCy with sentiment analysis
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("spacytextblob")

# Perform sentiment analysis on the poem
doc = nlp(poem_text)
polarity_score = doc._.blob.polarity

print(f"\nSentiment Analysis:")
print(f"Polarity score: {polarity_score}")
print(f"Range is [-1.0, 1.0] where -1.0 is most negative and 1.0 is most positive")

# Answer the question about positive/negative connotation
if polarity_score > 0:
    print("The poem has a more POSITIVE connotation.")
elif polarity_score < 0:
    print("The poem has a more NEGATIVE connotation.")
else:
    print("The poem is NEUTRAL.")
```

### 3. Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

```python
import requests
import json

def get_poem_and_save(author, poem_title, filename):
    """
    Function to get poem from Poetry DB API and save to JSON file
    Parameters:
    - author: poet's name
    - poem_title: title of the poem  
    - filename: name of file to save (without .json extension)
    """
    try:
        # Make API request
        url = f'https://poetrydb.org/author,title/{author};{poem_title}'
        response = requests.get(url)
        
        if response.status_code == 200:
            result = response.json()
            
            # Save to file
            with open(f"{filename}.json", "w") as f:
                json.dump(result, f, indent=2)
            
            print(f"Successfully saved '{poem_title}' by {author} to {filename}.json")
            return True
        else:
            print(f"Error {response.status_code} for '{poem_title}' by {author}")
            return False
            
    except Exception as e:
        print(f"Error getting '{poem_title}' by {author}: {e}")
        return False

# Test the function with 4 different poems
poems = [
    ("William Shakespeare", "Sonnet 18", "shakespeare_sonnet"),
    ("Edgar Allan Poe", "Annabel Lee", "poe_annabel"),
    ("Edgar Allan Poe", "The Raven", "raven_poem"),
    ("Edgar Allan Poe", "The Bells", "poe_bells")
]

for author, title, filename in poems:
    get_poem_and_save(author, title, filename)
```

### 4. Write a function that takes the name of a file that contains song lyrics, loads the file, performs sentiment analysis, and returns the polarity score.  Use this function to print the polarity scores (with the name of the song) of the three files you created in question 3.  Does the reported polarity match your understanding of the song's lyrics? Why or why not do you think that might be?  Answer the questions in either a comment in the code cell or a markdown cell under the code cell.

```python


# Set up spaCy (only need to do this once)
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("spacytextblob")

def analyze_poem_sentiment(filename):
    """
    Function to load a poem file and return its sentiment polarity score
    """
    try:
        with open(filename, "r") as f:
            poem_data = json.load(f)
        
        poem_text = '\n'.join(poem_data[0]['lines'])
        
        doc = nlp(poem_text)
        polarity_score = doc._.blob.polarity
        
        return polarity_score
        
    except Exception as e:
        print(f"Error analyzing {filename}: {e}")
        return None

# Test the function on the files created in Question 3
files_to_analyze = ["shakespeare_sonnet.json", "poe_annabel.json", "raven_poem.json", "poe_bells.json"]

print("Sentiment Analysis Results:")

for filename in files_to_analyze:
    polarity = analyze_poem_sentiment(filename)
    if polarity is not None:
        print(f"{filename}: {polarity:.3f}")
        
        # Interpret the score
        if polarity > 0.1:
            sentiment = "Positive"
        elif polarity < -0.1:
            sentiment = "Negative"  
        else:
            sentiment = "Neutral"
        print(f"   {sentiment} sentiment")
        print()
```
